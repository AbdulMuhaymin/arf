```python
import matplotlib.pyplot as plt

# Initialize lists to store steps and gradient errors
steps = []
gradient_errors = []

# Open and read the file
with open('error.dat', 'r') as file:
    lines = file.readlines()

# Parse the lines
for i in range(1, len(lines), 2):  # Start from the second line and skip every other line
    line = lines[i]
    gradient_error = float(line.split('=')[-1].split()[0])  # Split by '=' and take the last part, then split by space and take the first part
    gradient_errors.append(gradient_error)
    steps.append(i // 2)  # Add the step number

# Plot the data
plt.plot(steps, gradient_errors)
plt.ylim(0,0.001)
plt.xlabel('Step')
plt.ylabel('Gradient Error')
plt.title('Gradient Error over Steps')
plt.show()
```
